# Check if negative couplings lead to unphysical results for torsions, computing the functions for each set of atoms

## Functions and classes

```python
from __future__ import print_function, division
import time
import torch
import pickle
from tqdm.notebook import tqdm
from IPython.utils import io
import torch.nn as nn
from torch.nn import Parameter
import LocalEnergyVct as le
import numpy as np
from torch.utils.data import DataLoader, random_split
from data_classes import RNASeqDataset, LocalEnergyOpt
from sklearn.metrics import r2_score
# from my_script import get_target, loss_fn, train, test
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
def create_torsion_dict(ttype):
    tdict = {}
    for i in range(len(ttype)):
        key = str(ttype[i,:4].tolist())
        if key in tdict:
            tdict[key].append(i)
        else:
            tdict[key] = [i]
    return tdict       


def torsion_functions(ttype,phi_type,tpar,mult,datpar):
    tdict = create_torsion_dict(ttype)
    x = np.linspace(-np.pi,np.pi,100)
    rows_list = tdict.values()
    y = []
    for rows in rows_list:
        en = 0.
        for row in rows:
            idx = ttype[row,4]
            en += datpar[9] * datpar[28+phi_type[row]] * tpar[idx,0] * (1+torch.cos(mult[idx]*x - tpar[idx,1]))
        y.append(en)   
    return x, y, tdict.items()


def extract_torsion(dataset):
    
    lengths = dataset['lengths']
    features = dataset['features']
    ttypes = []
    phis = []

    for i in range(len(features)):
        atom = features[i,:lengths[i,0],0].squeeze().long()    
        ttype = features[i,:lengths[i,8],8].squeeze().long().view(-1,5)
        ttype[:,:4] = (ttype[:,:4]/3).long()
        ttype[:,4] -= 1    
        phi = le.assign_phitype(atom,ttype[:,0],ttype[:,1],ttype[:,2],ttype[:,3])
        ttypes.append(ttype)
        phis.append(phi)
        
    return ttypes, phis


def torsion_engine(state_dict):

    dataset = RNASeqDataset().dataset  
    model = LocalEnergyOpt()
    model.load_state_dict(torch.load("data/Results/" + state_dict))
    datpar = model.opt_pars.data
    tpar = model.tor_type.data
    tpar[13][0] = -0.2  
    tpar[14][0] = -0.3  # parameters that tendto become negative
    mult = model.multiplicity
    print(tpar)
    
    ttypes, phis = extract_torsion(dataset)
    print("Torsions extracted")
    
    y_list = []
    items_list = []
    for i in range(len(ttypes)):
        x, y, items = torsion_functions(ttypes[i],phis[i],tpar,mult,datpar)
        y_list.append(y)
        items_list.append(items)
        if i % 500 == 0 and i != 0:
            print(f"{i}/{len(ttypes)} functions computed")
    print("y_list and items_list created")
    
    return x, y_list, items_list
    
```

The analysis was carried out to check if the parameters that during the optimization tend to become negative for a sufficiently high learning rate appear only in composed functions, or also on their own. If the case was the former, then even a negative value would have been acceptable as long as the total function was globally positive.

Unfortunately, plots reveal that the considered terms appear also on their own, therefore a negative value is unphysical and the optimization must be tuned to avoid such case. Anyway, even if not negative, their values decrease very close to 0, however the parameters are initialized
