## RMSD plots

```python
import mdtraj as md
import numpy as np
import sys
import subprocess
import math
import pylab
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
from matplotlib import rc
import matplotlib.backends.backend_pdf
from os.path import exists
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

plt.style.use('default')
plt.rcParams["font.family"] = "serif"
rc('text.latex', preamble=r'\usepackage{cmbright}')
```

```python
# compute rmsd between relaxed structure and AMBER initial structure. A better version is already present
# in notebook relaxation_analysis.ipynb

with open('relaxation_names.txt','r') as f:
    file_list = f.read().splitlines()
    # file_list_amber = [line.split('_',maxsplit=2)[1] for line in file_list]

rmsd_PSGD = []
for i,file in enumerate(file_list):
    
    file_amber = file.split('_',maxsplit=1)[1]  
    
    if not exists('Amber/Minimized/'+file_amber+'.pdb'):
        continue
    if not exists('HiRE_old/Relaxed_conformations/relaxed_'+file_amber+'.pdb'):
        continue
    
    print(file_amber+'     %i/%i' %(i+1,len(file_list)))
    ref = md.load('Amber/Minimized/'+file_amber+'.pdb')
    res = md.load_frame('HiRE_old/Relaxed_conformations/relaxed_'+file_amber+'.pdb',-1)
    top = ref.topology
    
    rmsd_tmp = md.rmsd(res,ref) 
    rmsd_PSGD.append(rmsd_tmp)
rmsd_PSGD = np.array(rmsd_PSGD)
    
np.save('HiRE_old/rmsd.npy',rmsd_PSGD)
```

```python
# load rmsd for different relaxations, convert to Angstroms and plot distributions
rmsd_free = np.load('FreeMin/rmsd.npy')*10
rmsd_psgd = np.load('PSGD/rmsd.npy')*10
rmsd_old = np.load('HiRE_old/rmsd.npy')*10
rmsd_nogeom = np.load('NoGeom/rmsd.npy')*10
rmsd_tanh = np.load('Tanh/rmsd.npy')*10
rmsd_dof = np.load('Less_dof/rmsd.npy')*10
rmsd_free_new = np.load('FreeMin_newangles/rmsd.npy')*10

fig = plt.figure(figsize=(10,7))
sns.distplot(x=rmsd_old, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='HiRE old', color='steelblue')
sns.distplot(x=rmsd_free, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='Free Min', color='firebrick')
'''
sns.distplot(x=rmsd_nogeom, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='NG')
sns.distplot(x=rmsd_psgd, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='PSGD')
sns.distplot(x=rmsd_tanh, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='Tanh')
sns.distplot(x=rmsd_dof, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='Less dof')
'''
sns.distplot(x=rmsd_free_new, hist = False, kde = True,
        kde_kws = {'shade': True, 'linewidth': 2}, label='Free Min new')
plt.legend(fontsize=15)
plt.grid(linewidth=0.3)
plt.xlabel(r'RMSD ($\AA$)', fontsize=18)
plt.ylabel('Density',fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=13)

plt.savefig('Images/rmsd_distribution.png',bbox_inches='tight',dpi=100)
```

```python
# boxplot 

rmsd = np.concatenate((rmsd_old, rmsd_free, rmsd_psgd, rmsd_nogeom, rmsd_tanh, rmsd_dof, rmsd_free_new),axis=1)
print(rmsd.shape)

fig = plt.figure(figsize=(9,5))

medianprops = dict(color="black",linewidth=2)
boxprops = dict(color="black",linewidth=1.5)
meanprops = dict(color="black",linewidth=1.5,linestyle='--')
whiskerprops = dict(linewidth=1.2,linestyle='--')

plt.boxplot(rmsd, sym='', widths=0.7, meanline=True, medianprops=medianprops, 
            boxprops=boxprops, meanprops=meanprops, whiskerprops=whiskerprops)
locs, labels = plt.xticks()
plt.axhline(y=1, color='r', linestyle='--')
plt.xticks(locs,['HiRE old', 'Free min', 'PSGD', 'NG', 'Tanh', 'Less dof', 'Free min new']) #,rotation=45)
plt.tick_params(axis='x', which='major', labelsize=16)
plt.tick_params(axis='y', which='major', labelsize=12)
# plt.grid(axis='y',linewidth=0.2)
plt.ylabel(r'RMSD ($\AA$)', fontsize=16)

plt.savefig('rmsd_stat.png',bbox_inches='tight',dpi=100)
```
